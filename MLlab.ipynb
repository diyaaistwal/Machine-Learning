```python
#create linear regression
import numpy as np
import matplotlib.pyplot as plt
x=np.random.randint(0,100,10)
y=np.random.randint(0,100,10)
plt.scatter(x, y, color='blue')
m, b = np.polyfit(x, y, 1)
y_fit = m * x + b
plt.plot(x, y_fit, color='red')
plt.show()
```
